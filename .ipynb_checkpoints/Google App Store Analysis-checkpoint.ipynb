```python
# Preparing Data

# gps_df[gps_df['Installs'] == 'Free']
gps_df.drop(gps_df[(gps_df['Installs'] == 'Free')].index, inplace=True)
gps_df.drop(gps_df[(gps_df['Category'] == '1.9')].index, inplace=True)

def parse_installs(x):
    try:
        result = int(str(x).replace(',', '').strip('+')) if str(x) else 0
    except:
        result = 0
    return result

gps_df['Installs_in_num'] = gps_df['Installs'].apply(parse_installs)
gps_df.head(5)
```

```python
gps_df[(gps_df['Rating'] > 4.0) & (gps_df['Installs_in_num'] > 1000000)]
```
